```python
import math

infasta = '/content/drive/MyDrive/Abf1_sf_90_fps_with_sf_fimo.fasta'

def prep_fasta(fasta, combine):
  
  ff = open(infasta) # A set of FASTA sequences, no size limit (whole genomes acceptable)
  rawfa = ff.read().splitlines()
  fastseq = []
  
  for line in rawfa:
    if not (">" in line):
      fastseq.append(str(line)) # Trims FASTA headers
  
  if combine == "TRUE":
    fastseq = "".join(fastseq)
  
  return fastseq

def build_kmers(dnalist, k, unique):
  # Converts a list of DNA sequence strings into their component k-mer strings
  # Reports a dictionary of the input sequence, its kmers, and their counts
  
  merdict = {}
  
  for seq in dnalist:
    
    inseq = str(seq)
    kmercount = {}
    mls = []
    
    for i in range((len(inseq) - 1) - k):
      mer = inseq[i:i+k] # Report the kmers found in the input string
      mls.append(mer) # Add the kmer to the list
    
    for km in mls:
      if not (km in kmercount): # Check if current kmer exists in dictionary
        kmercount[km] = 0 # if not, add it as a new entry
      kmercount[km] += 1 # increment the counter for the k-mer
    
    if not (inseq in merdict): #Have we already got a k-mer dictionary for this sequence? If not, add this one
      merdict[inseq] = kmercount
  
  return merdict

newseqs = prep_fasta(infasta, "FALSE")

outmers = build_kmers(newseqs, 2, "TRUE")
with open('mercount.kcf', "w+") as newout:
  print(outmers, file = newout)


```
